```python
def find_task_relation(task1, task2, kernel_map=None):
    """Quantifies the relation between two tasks."""
    global model_name
    if isinstance(task1, list):
        pass
    elif isinstance(task1, str):
        task1 = [task1]
    if isinstance(task2, list):
        pass
    elif isinstance(task2, str):
        task2 = [task2]
    tasks = task1 + task2
    kernels = []
    n1 = 0
    for i, task in enumerate(tasks):
        if kernel_map is None or task not in kernel_map:
            model_path = f'saved_models/{model_name}/{task}.h5'
            model = keras.models.load_model(model_path, compile=False, custom_objects={'PositionEmbedding': PositionEmbedding})
            kernels.append(model.layers[-1].kernel)
        else:
            kernels.append(kernel_map[task])
        if i < len(task1):
            n1 += kernels[-1].shape[1]
    kernel = np.concatenate(kernels, axis=1)
    n1 = n1
    n2 = kernel.shape[1] - n1
    
    u, s, vh = np.linalg.svd(kernel, full_matrices=False, hermitian=False)
    
    # effective_rank = s.shape[0]
    # effective_rank_fraction = 0.99
    # for i in range(s.shape[0]):
    #     if np.sum(s[i:] ** 2) < (1.0 - effective_rank_fraction) * np.sum(s ** 2):
    #         effective_rank = i
    #         break

    # if effective_rank == s.shape[0]:
    #     effective_rank -= 1

    # norm1 = np.linalg.norm(np.matmul(np.transpose(vh[:effective_rank, :n1]), vh[:effective_rank, :n1]), 2)
    # norm2 = np.linalg.norm(np.matmul(np.transpose(vh[:effective_rank, n1:]), vh[:effective_rank, n1:]), 2)
    # cross = np.linalg.norm(np.matmul(np.transpose(vh[:effective_rank, n1:]), vh[:effective_rank, :n1]), 2)
    # return cross / np.sqrt(norm1 * norm2)

    # Experimental
    u1, s1, _ = np.linalg.svd(kernel[:, :n1], full_matrices=False, hermitian=False)
    u2, s2, _ = np.linalg.svd(kernel[:, n1:], full_matrices=False, hermitian=False)

    effective_rank_1 = u1.shape[1]
    effective_rank_fraction = 0.25
    for i in range(s1.shape[0]):
        if np.sum(s1[i:] ** 2) < (1.0 - effective_rank_fraction) * np.sum(s1 ** 2):
            effective_rank_1 = i
            break
    effective_rank_2 = u2.shape[1]
    for i in range(s2.shape[0]):
        if np.sum(s2[i:] ** 2) < (1.0 - effective_rank_fraction) * np.sum(s2 ** 2):
            effective_rank_2 = i
            break
    coeff = np.matmul(np.transpose(u1[:, :effective_rank_1]), u2[:, :effective_rank_2])
    return np.linalg.norm(coeff, 2)
```

```python
# In this colab, we calculate the low-rank projection matrix for any group of tasks.
relation_matrix = np.zeros(shape=(len(tasks), len(tasks)))
for i in range(len(tasks)):
    for j in range(i, len(tasks)):
        relation_matrix[i, j] = find_task_relation(tasks[i], tasks[j], kernel_map)
        relation_matrix[j, i] = relation_matrix[i, j] 
```

```python
relation_matrix_two_channel = relation_matrix[:-3, :-3]

with plt.style.context(spstyle.get_style('nature-reviews')):
    # Generate a sample n x n matrix
    n = len(relation_matrix_two_channel[:, 0])
    matrix = relation_matrix_two_channel

    plt.figure(figsize=(12, 8))
    # Create the confusion matrix plot
    fig, ax = plt.subplots()
    
    # Set the colormap
    cmap = plt.cm.Blues
    
    # Create the heatmap
    im = ax.imshow(matrix, interpolation='nearest', cmap=cmap)
    
    # Add colorbar
    fig.colorbar(im, label='Alignment')
    
    # Set axis labels
    # ax.set_xlabel('Tasks')
    # ax.set_ylabel('Tasks')
    
    # Set tick labels
    ax.set_xticks(np.arange(n))
    ax.set_yticks(np.arange(n))
    ax.set_xticklabels(names[:n], rotation=90)
    ax.set_yticklabels(names[:n])
    
    # Add text annotations
    for i in range(n):
        for j in range(n):
            text = ax.text(j, i, np.round(matrix[i, j], 2),
                          ha='center', va='center', fontsize=5,
                          bbox=dict(boxstyle='round', facecolor='white', edgecolor='1.0'))
    
    # Set title
    ax.set_title('Relation Matrix')
    
    # Show the plot
    plt.tight_layout()
    # plt.show()
    plt.savefig('figures/fig5-relation-matrix.svg', transparent=True, dpi=200)
```

```python
def create_bar_plot(data, label):
    x_labels = label
    
    # Create the bar plot
    plt.figure(figsize=(8, 6))
    plt.bar(x_labels, data)
    
    # Customize the x-axis ticks
    plt.xticks(rotation=0)  # Rotate and right-align x-axis labels
    plt.xlabel("Tasks")
    plt.ylabel("Relation")
    # plt.title("Bar Plot with Custom X-axis Ticks")
    
    # Show the plot
    plt.tight_layout()
    # plt.show()
```
